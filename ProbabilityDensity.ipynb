```python
#Import necessary library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Export dat file

filename = 'psi3.dat'
data = np.loadtxt(filename)

#Make lists to store x and psi value
x = []
psi = []
psi2 = []

for i in range(len(data)):
  x.append(data[i][0])
  psi.append(data[i][1])
  psi2.append(abs(data[i][1])**2)

print(x)
print(psi)
```

```python
plt.plot(x, psi2)
```

```python
def integrate(xpos, psi2, xlim):
  '''
  Calculates the probability using Simpson 1/3 integration method
  Input   :X data, psi^2 data, x boundary
  Return  : probability
  '''

  xmin = min(xpos)
  h = abs(x[2]-x[1])/2
  dx = abs(x[2]-x[1])
  prob1, prob2 = 0, 0
  
  for i in range(len(xpos)-2):
    #Probability of the first region
    if xmin <= xlim:
      prob1 = prob1 + (psi2[i] + 4*psi2[i+1] + psi2[i+2])
      xmin = xmin + dx

    #Probability of the second region
    else:
      prob2 = prob2 + (psi2[i] + 4*psi2[i+1] + psi2[i+2])
      xmin = xmin + dx

  prob1 = prob1 * h/3
  prob2 = prob2 * h/3

  return prob1, prob2
```

```python
#Calculates probability

xlim = 0.5

prob1, prob2 = integrate(x, psi2, xlim)
print(prob1)
print(prob2)
print(prob1+prob2)
```
